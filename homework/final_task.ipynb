get_dummies?


```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
data.loc[data["whoAmI"] == "human" , "human"] = "I_Am_A_Human"
data.loc[data['whoAmI'] == "robot", "robot"] = "I_Am_A_Toster"
data.loc[(data["whoAmI"] == "robot") & (data["robot"] == 'I_Am_A_Toster'), "human"] = "i_am_not_a_Human"
data.loc[(data["whoAmI"] == "human") & (data["human"] == 'I_Am_A_Human'), "robot"] = "i_am_NOT_a_Toster"

data
```
